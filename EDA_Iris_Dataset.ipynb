# Dataset Description
This dataset contains measurements of Iris flowers with the following columns:

1. Id: A unique identifier for each record.
2. SepalLengthCm: Length of the sepal in centimeters.
3. SepalWidthCm: Width of the sepal in centimeters.
4. PetalLengthCm: Length of the petal in centimeters.
5. PetalWidthCm: Width of the petal in centimeters.
6. Species: The species of the Iris flower, with all entries being "Iris-setosa".

### Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Import Dataset

```python
df=pd.read_csv("Iris.csv")
```

### Visualization

#### countplot

```python
sns.countplot(x='Species', data=df, )
plt.title('Count Plot of Iris Species', fontsize=15)
plt.xlabel('Species', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```

##### scatterplot

```python
sns.scatterplot(x='SepalLengthCm', y='SepalWidthCm',hue='Species', data=df, )

plt.legend(bbox_to_anchor=(1, 1))

plt.show()
```

#### scatterplot

```python
sns.scatterplot(x='PetalLengthCm', y='PetalWidthCm',hue='Species', data=df, )

plt.legend(bbox_to_anchor=(1, 1))

plt.show()
```

#### Boxplot

```python
sns.boxplot(x='SepalWidthCm', data=df)
```

#### histograms

```python
plt.figure(figsize=(10, 6))
x=df.drop('Id', axis=1)
x.hist(bins=15, figsize=(15, 10), layout=(2, 3))
plt.suptitle('Histograms of Iris Dataset Features')
plt.show()
```

#### heatmap : corelation in numerical values

```python
plt.figure(figsize=(10, 6))
x=df.drop(columns=['Id','Species'])
x=sns.heatmap(x.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Heatmap of Feature Correlations')
plt.show()
```

#### Pairplot

```python
sns.pairplot(df, hue='Species')
plt.suptitle('Pairplot of Iris Dataset',)
plt.show()
```

#### violin plot

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Species', y='SepalWidthCm', data=df)
plt.title('Violin Plot of Sepal Width by Species')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Species', y='SepalLengthCm', data=df)
plt.title('Violin Plot of Sepal Width by Species')
plt.show()
```

# Observations
1. Sepal lengths range from 4.6 cm to 5.1 cm, and sepal widths range from 3.0 cm to 3.6 cm.
2. Petal lengths are between 1.3 cm and 1.5 cm, and petal widths are uniformly 0.2 cm.
3. The heatmap reveals a strong positive correlation between sepal length and sepal width.
